```python
import pandas as pd
import numpy as np
```

```python
data = {
    "ticket_text": [
        "My internet is not working",
        "App keeps crashing when I open it",
        "Unable to login to my account",
        "Password reset link not working",
        "Website is very slow",
        "Server error 500 showing",
        "Payment failed but money deducted",
        "I was charged twice for my order",
        "Refund not received yet",
        "Incorrect billing amount",
        "Subscription cancelled but charged",
        "Delivery is delayed",
        "Order not delivered",
        "Received wrong product",
        "Tracking information not updating",
        "Cannot update profile details",
        "Account temporarily locked",
        "Email verification not received",
        "Discount not applied at checkout",
        "App not loading after update",
        "Unable to add items to cart",
        "Order confirmation email not received",
        "Payment gateway timeout error",
        "Login session expired quickly",
        "Product damaged on delivery"
    ],
    "category": [
        "Technical","Technical","Account","Account","Technical",
        "Technical","Billing","Billing","Billing","Billing",
        "Billing","Delivery","Delivery","Delivery","Delivery",
        "Account","Account","Account","Billing","Technical",
        "Technical","Delivery","Billing","Account","Delivery"
    ],
    "priority": [
        "High","High","Medium","Medium","Low",
        "High","High","High","Medium","Medium",
        "High","Low","High","Medium","Low",
        "Medium","High","Medium","Medium","High",
        "Medium","Low","High","Medium","High"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df["ticket_text"])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df["category"])

y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train model
model = MultinomialNB()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
new_ticket = ["I cannot login and my account is locked"]
new_vector = vectorizer.transform(new_ticket)

prediction = model.predict(new_vector)

label_encoder.inverse_transform(prediction)
```

```python
from sklearn.preprocessing import LabelEncoder

priority_encoder = LabelEncoder()
y_priority = priority_encoder.fit_transform(df["priority"])

y_priority
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split data
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X, y_priority, test_size=0.3, random_state=42
)

# Train model
priority_model = MultinomialNB()
priority_model.fit(X_train_p, y_train_p)

# Predict
y_pred_p = priority_model.predict(X_test_p)

# Accuracy
priority_accuracy = accuracy_score(y_test_p, y_pred_p)
priority_accuracy
```

```python
new_ticket = ["Payment failed and money deducted"]

new_vector = vectorizer.transform(new_ticket)

priority_prediction = priority_model.predict(new_vector)

priority_encoder.inverse_transform(priority_prediction)
```

```python
new_ticket = ["Payment failed and money deducted"]

# Transform text
new_vector = vectorizer.transform(new_ticket)

# Predict category
category_prediction = model.predict(new_vector)

# Predict priority
priority_prediction = priority_model.predict(new_vector)

# Print clean output
print("Ticket:", new_ticket[0])
print("Predicted Category:", label_encoder.inverse_transform(category_prediction)[0])
print("Predicted Priority:", priority_encoder.inverse_transform(priority_prediction)[0])
print()
print("Category Model Accuracy:", accuracy)
print("Priority Model Accuracy:", priority_accuracy)
```
